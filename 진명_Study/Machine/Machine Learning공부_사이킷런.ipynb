# model_selection 모듈
* 학습용 데이터와 테스트 데이터로 분리시키거나 할때 사용되는 다양한 메소드 제공
* 교차 검증 분할 및 평가 관련된 메소드 제공
* estimator의 하이퍼 파라미터 튜닝을 위한 다양한 함수와 클래스 제공

###  ✅ train_test_split() : 학습(train) / 테스트(test) 데이터 세트 분리

```python
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

diabetes = load_diabetes()

X_train,X_test, y_train,y_test = train_test_split(diabetes.data,diabetes.target,test_size=0.3)

# train 70% test 30%
# return 4개 나옴
# 대문자 X 여러개의 feature가 포함되어있어서
# target이 되는 y값 벡터형태이다.
# 하나의 벡터로 구성되어있을대는 소문자
# 매트리스에 대한 데이터를 담고 있을때는 대문자

model = LinearRegression() # 인스턴스화
model.fit(X_train,y_train) # 데이터를 실질적으로 학습하기 위해, 학습용 데이터를 가지고 모델을 학습

# 지금 현재의 모델이 훈련을 함으로써 나오는 점수
print("학습 데이터 점수 : {}".format(model.score(X_train,y_train)))
print("평가 데이터 점수 : {}".format(model.score(X_test,y_test)))
```

```python
import matplotlib.pyplot as plt

# predict를 통해 실제적으로 예측을 했는지 새로운 데이터에 대한 예측결과가나옴

predicted = model.predict(X_test)
expected = y_test
plt.figure(figsize=(8,4))
plt.scatter(expected,predicted)
plt.plot([30,350],[30,350],'--r')
plt.tight_layout() 
# 빨간 선에 가까울수록 잘 맞춘것, 가깝지 않은게 많다.
# 잘 맞추려면 모델을 다른걸 쓰거나~, 데이터 전처리를 통해 개선
```

###  ✅ cross_val_score() : 교차 검증

```python
from sklearn.model_selection import cross_val_score,cross_validate

scores = cross_val_score(model,diabetes.data, diabetes.target, cv=5)
# 몇개로 나눌지 값 넣어줘야함 cv

print("교차 검증 정확도 : {}".format(scores))
print("교차 검증 정확도 : {} +/- {}".format(np.mean(scores),np.std(scores)))
```

###  ✅ GridSearchCV() : 교차 검증과 최적 하이퍼 파라미터 찾기
* 훈련 단계에서 학습한 파라미터에 영향을 받아서 최상의 파라미터를 찾는 일은 항상 어려운 문제
* 다양한 모델의 훈련 과정을 자동화, 교차 검사를 사용해 최적 값을 제공하는 도구 필요
* cross validation을 여러번 반복해서 최적의 하이퍼 파라미터를 찾는 방식

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Ridge
import pandas as pd

alpha = [0.001, 0.01, 0.1, 1, 10, 100, 1000]
param_grid = dict(alpha=alpha)
gs = GridSearchCV(estimator=Ridge(), param_grid=param_grid, cv= 10)
result = gs.fit(diabetes.data, diabetes.target)

print("최적 점수 : {}".format(result.best_score_))
print("최적 파라미터 : {}".format(result.best_params_))
print(gs.best_estimator_)
pd.DataFrame(result.cv_results_)

# 알파값으로 7번 돌림
# 최적의 값은 0.1로 돌리는게 최적 파라미터
# 최고 에스터메이터는 0.1
```

* multiprocessing을 이용한 GridSearchCV

```python
import multiprocessing
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression

iris = load_iris()

param_grid = [{'penalty':['l1','l2'],
               'C' : [ 1.5, 2.0, 2.5, 3.0, 3.5 ]}]

gs2 = GridSearchCV(estimator=LogisticRegression(), param_grid = param_grid,
                  scoring='accuracy', cv=10, n_jobs = multiprocessing.cpu_count())

result = gs2.fit(iris.data, iris.target)

print("최적 점수 : {}".format(result.best_score_))
print("최적 파라미터 : {}".format(result.best_params_))
print(gs2.best_estimator_)
pd.DataFrame(result.cv_results_)
```

### 🟨 StandardScaler : 표준화 클래스
* 각 특성의 평균을 0, 분산을 1로 스케일링합니다. 즉 데이터를 정규분포로 만듭니다. 하한값과 상한값이 존재하지 않을 수 있기에, 어떤 알고리즘에서는 문제가 있을 수 있습니다. 회귀보다 분류에 유용합니다.

```python
iris = load_iris()
iris_df = pd.DataFrame(data = iris.data, columns = iris.feature_names)
iris_df.describe()
```

```python
X_train, X_test, y_train,y_test = train_test_split(iris_df_scaled,
                                                   iris.target,
                                                   test_size=0.3)
model = LogisticRegression()
model.fit(X_train,y_train)

print("훈련 데이터 점수 : {}".format(model.score(X_train,y_train)))
print("평가 데이터 점수 : {}".format(model.score(X_test,y_test)))

# 점수가 꽤 높은데?
```

# 💜성능 평가 지표

### 정확도(Accuracy)
* 정확도는 전체 예측 데이터 건수 중 예측 결과가 동일한 데이터 건수로 계산
* 사이킷 런에서는 accuracy_score함수를 제공

```python
from sklearn.datasets import make_classification
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

X,y = make_classification(n_samples=1000,
                          n_features=2,
                          n_informative=2,
                          n_redundant=0,
                          n_clusters_per_class=1) # 데이터를 만들어내는 함수

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

model = LogisticRegression()
model.fit(X_train,y_train)

print("훈련 데이터 점수 : {}".format(model.score(X_train, y_train)))
print("평가 데이터 점수 : {}".format(model.score(X_test, y_test)))

# 정확도는 한번 맞춰봐야한다.
# 실제 테스트 데이터에 대해서 얼만큼 잘 예측을 했는지에 대한 결과
predict = model.predict(X_test)
print("정확도 : {}".format(accuracy_score(y_test,predict)))
```
