```python
import requests
import xml.etree.ElementTree as ET
import pandas as pd
```

```python
url ='http://openapi.data.go.kr/openapi/service/rest/Covid19/getCovid19InfStateJson?serviceKey=WRvGBkn9UEtw%2BAsg0tYo210Etxvb1QEcAX%2BwfWvOxVGYJkh1CNZ%2FY4QFa0r7j4bhT4NjPu7z1i1ck8ZgsKMt2Q%3D%3D&pageNo=1&numOfRows=10&startCreateDt=20210310&endCreateDt=20210515'
```

```python
r = requests.get(url)
root = ET.fromstring(r.text)
items = root.iter(tag = 'item')
```

```python
total = []
for element in items:
    tmp_dict = dict()
    for x in element:
        tmp_dict[x.tag] = x.text
    total.append(tmp_dict)
```

```python
df = pd.DataFrame(total)
```
